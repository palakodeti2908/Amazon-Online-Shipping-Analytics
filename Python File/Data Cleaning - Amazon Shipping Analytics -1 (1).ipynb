```python
# Dropping the Ship region and ship Postal code column from orders table because its has alphanumeric values
orders_df = pd.read_csv(r"/Users/bharadwajpalakodeti/Documents/Bharadwaj/USA/NEU/Comp Viz/Amazon Shipping Analytics/Duplicate CSV's/orders1.csv")
orders_df
```

```python
orders_df = orders_df.drop(['shipRegion', 'shipPostalCode'], axis =1)
orders_df
```

```python
orders_df.info()
```

```python
orders_df['requiredDate'] = pd.to_datetime(orders_df.requiredDate)
orders_df
```

```python
orders_df['orderDate'] = pd.to_datetime(orders_df.orderDate)
orders_df
```

```python
u = orders_df.select_dtypes(include=['datetime'])
orders_df[u.columns] = u.fillna(pd.to_datetime('today'))
orders_df
```

```python
orders_df['shippedDate'] = pd.to_datetime(orders_df.shippedDate)
orders_df
```

#### Orders Table is now cleaned
